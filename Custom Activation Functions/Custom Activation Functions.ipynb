```python
import numpy as np
```

```python
x = np.linspace(-30,30, 10000)
```

### Sin Functions

```python
def func1(x):
  return x*np.sin(x)
```

```python
import matplotlib.pyplot as plt

plt.plot(x,func1(x),label = "xsinx", color = "Blue")
plt.xlabel("X")
plt.ylabel("X Sin(X)")
plt.grid(True)
plt.title("Function 1")
plt.legend()
plt.show()
```

```python
def func2(x):
  return x*np.sin(np.abs(x))  # x sin|x|
```

```python
plt.plot(x,func2(x),label="xsin|x|",color="blue")
plt.xlabel("X")
plt.ylabel("X Sin|X|")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def func3(x): return np.abs(x) * np.sin(x)
```

```python
plt.plot(x,func3(x), label = "|X| SinX")
plt.legend()
plt.xlabel("X")
plt.legend()
plt.grid(True)
```

### Ghost ReLU and others

```python
import numpy as np
x = np.linspace(-100,100,10000)

func_1 = lambda x: x/np.tanh(x)

import matplotlib.pyplot as plt
plt.plot(x,func_1(x),label = "X/Tanh(X)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
func_2 = lambda x: 1/(1 + np.exp(-np.tanh(x)))

plt.plot(x,func_2(x), label = "Sigmoid(Tanh(X))")
plt.grid()
plt.legend()
plt.show()
```

```python
# Ghost ReLU!

func_3 = lambda x: x/(1 + np.exp(-np.tanh(x)))

plt.plot(x,func_3(x), label = "X Sigmoid(Tanh(X))")
plt.grid()
plt.legend()
plt.show()
```

```python
func_3 = lambda x: x * np.tanh(1/(1+np.exp(-x)))

plt.plot(x,func_3(x), label = "X Tanh(Sigmoid(X))")
plt.grid()
plt.legend()
# looks like relu, i name it Soft ReLU
```

```python
# sigmoid(x) = 0.85, sigmoid(x) = 0.15, x=??? ..... x = 1.735, -1.735
# we need a function that squishes values to [-1.735,1.735].... = 1.735 tanh
# so sigmoid(1.735tanh(x)), -ve x will be scaled by 0.15, +ve x by 0.85
# gammoid (random name)

func_4 = lambda x: 1/(1+np.exp(-1.735*np.tanh(x)))

plt.plot(x,func_4(x), label = "Sigmoid(1.735 Tanh(X))")
plt.legend()
plt.grid()
plt.show()
```

```python
# new Ghost ReLU
func_4 = lambda x: x/(1+np.exp(-2.4*np.tanh(x))) # we can make it like x*sigmoid(a*tanh(x)) where a is a learnable parameter

plt.plot(x,func_4(x), label = "X Sigmoid(4 Tanh(X))")
plt.legend()
plt.grid()
plt.show()
```

```python
e = 0.000001
a = 10
func_5 = lambda x: x*np.log(a*(e + (1/(1+np.exp(-x)))))

plt.plot(x,func_5(x),label = "X Log(Sigmoid(X))")
plt.grid()
plt.legend()
plt.show()
```

```python
j = 1.025
func_6 = lambda x: x/(1+j**(-x)) # checking sigmoid and tanh with any other constant than e

plt.plot(x,func_6(x), label=(f"X Sigmoid(X) with {j} in place of e"))
plt.grid()
plt.legend()
plt.show()
```

```python
j=1.05
func_7 = lambda x: (j**x - j**(-x))/(j**x + j**(-x))

plt.plot(x,func_7(x))
plt.grid()
```

```python
import numpy as np
x = np.linspace(-100,100,10000)

def sigmoid(x):
  return 1/(1+np.exp(-x))

import matplotlib.pyplot as plt
plt.plot(x,x*sigmoid(sigmoid(x)),label = "Sigmoid(Sigmoid(X))")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(x,np.sqrt(np.abs(x))*x/(np.abs(x))*sigmoid(sigmoid(x)), label = "Sign of x * sqrt |x| * Sigmoid(Sigmoid(X))")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def arcsinh(x):
  return np.log(x+np.sqrt(x**2 + 1))

plt.plot(x,arcsinh(x))
plt.grid()
```

```python
x = np.linspace(-4,4,1000)
plt.plot(x,x*sigmoid(1.735*np.tanh(x)), label = "Ghost ReLU")
plt.grid()
plt.legend()
plt.show()
```
