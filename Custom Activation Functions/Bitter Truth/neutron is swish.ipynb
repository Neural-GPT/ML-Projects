```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigmoid = lambda x: 1/(1+np.exp(-x))
neutron = lambda x: x*np.exp(x)/(1+np.exp(x))
```

```python
x = np.linspace(-10,10,1000)

fig, axis = plt.subplots(1,2,figsize=(12,6))

axis[0].plot(x,x*sigmoid(x),label="Swish")
axis[0].legend()
axis[0].grid()

axis[1].plot(x,neutron(x),label="Neutron")
axis[1].legend()
axis[1].grid()

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-25,25,2000)

a = 1.2
atomic_swish = lambda x: x*a**x/(1+a**x)

plt.plot(x,atomic_swish(x),label="Atomic Swish")
plt.grid()
plt.legend()
```

```python
modified_xtanh = lambda x: x*(np.tanh(x))**4 # because tanh gave -ve output for -ve x so gating with x it gave +ve output for -0ve x which is symmetric, lets test this

plt.plot(x,modified_xtanh(x),label="modified x tanh")
plt.grid()
plt.legend()
```
