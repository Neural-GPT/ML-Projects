```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigmoid = lambda x: 1/(1+np.exp(-x))
gammoid1 = lambda x: x/(1+np.abs(x))
grelu1 = lambda x: x*sigmoid(gammoid1(x))
```

```python
x = np.linspace(-15,15,10000)
```

```python
plt.plot(x,gammoid1(x), label = "X/(1+|X|)")
plt.grid()
plt.legend() #bro its so good, continuos and differentiable , outputs from -1 to 1 , looks just like tanh but more efficient!
```

```python
gammoid2 = lambda x: x/(1+x)
gammoid3 = lambda x: x/(x**2 + 1)**0.25
gammoid4 = lambda x: x/(x**2+1)**0.5
funcs = [gammoid2, gammoid3, gammoid4]

for func in funcs:
  plt.plot(x,func(x),label = f"{func}")
  plt.grid()
  # plt.legend()
  plt.tight_layout()
  plt.show()
```

```python
plt.plot(x,np.tanh(x))

plt.grid()
```

```python
a = 1.735
grelu1 = lambda x: x*sigmoid(a*gammoid1(x))
grelu3 = lambda x: x*sigmoid(a*gammoid3(x))
grelu4 = lambda x: x*sigmoid(a*gammoid4(x))

funcs = [grelu1, grelu3, grelu4]

for func in funcs:
  plt.plot(x,func(x))
  plt.grid()
  # plt.legend()
  plt.tight_layout()
  plt.show()
```
